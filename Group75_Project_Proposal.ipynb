Exploring the Study Time, Repetition Amount, and Exam Performance Values for Predicting the Knowledge Level of Students: A Data-driven Analysis

Introduction:

In order to better understand how much a student was able to understand the content of a subject, it is possible to analyze several parameters that involve their learning during the course. User Knowledge Modelling involves capturing and representing a user's knowledge, skills, and expertise in a specific domain. This project aims to predict a user's knowledge level using the STG, SCG, and PEG values. We will utilise the "User Knowledge" dataset, obtained from the UCI Machine Learning Repository, which focuses on students' knowledge levels in Electrical DC Machines. The dataset consists of 258 rows and 4 columns, including the classification of user knowledge. We begin by cleaning and preprocessing the dataset to ensure its integrity and quality. Next, we employ feature selection techniques to identify the most relevant variables, followed by model training using the K-Nearest Neighbors algorithm.

Preliminary exploratory data analysis:

```r
library(readxl)
library(tidyverse)
```

```r
#the second sheet presents the training data and the third sheet presents the testing data
knowledge_training <- read_excel(path = "Data_User_Modeling_Dataset_Hamdi Tolga KAHRAMAN.xls",sheet = 2)
knowledge_testing <- read_excel(path = "Data_User_Modeling_Dataset_Hamdi Tolga KAHRAMAN.xls",sheet = 3)
```

```r
#first, we selected the variables
knowledge_training <- knowledge_training |>
            select(STG,SCG, PEG, UNS)

knowledge_testing <- knowledge_testing |>
            select(STG,SCG, PEG, UNS)

#then we renamed the columns names
knowledge_training <- knowledge_training |>
        rename(
            study_time = STG,
            repetition_number = SCG,
            exam_performance = PEG,
            knowledge_level = UNS)

knowledge_testing <- knowledge_testing|>
        rename(
            study_time = STG,
            repetition_number = SCG,
            exam_performance = PEG,
            knowledge_level = UNS)

#finally, since knowledge_level have a limited number of values, we changed it to a factor data type
knowledge_training <- knowledge_training |>
        mutate(knowledge_level = as_factor(knowledge_level))

knowledge_testing <- knowledge_testing|>
        mutate(knowledge_level = as_factor(knowledge_level))
```

```r
options(repr.plot.width = 25, repr.plot.height = 8)

knowledge_plots <- knowledge_training |>
    ggplot(aes(x = study_time, y = repetition_number, color = exam_performance)) +
    geom_point(size = 3) +
    labs(colour = "Exam performance\n(standardized)")+
    xlab("Study Time (standardized)") + 
    ylab("Amount of Repetition (standardized)") +
    facet_grid(~factor(knowledge_level, levels = c("very_low", "Low", "Middle", "High"))) +
    ggtitle("The knowledge status of students about the subject of Electrical DC Machines") +
    theme(text = element_text(size=20))


knowledge_plots
```
